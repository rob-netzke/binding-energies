#### data processing

```python
from PyPDF2 import PdfReader
import pandas as pd

reader = PdfReader("data/binding_energies.pdf")
pages = reader.pages
rows = []

for ind,page in enumerate(pages):
  print(ind)
  print(page.extract_text().replace("\n", " "))
  if ind == 0:
    rows.append(page.extract_text().replace("\n", " ").replace("2 H", "2H").replace("1 H", "1H").replace("3 H", "3H").replace("TABLE 2Nuclear Binding Energies A Z BE (MeV)  A Z BE (MeV)  A Z BE (MeV)  A Z BE (MeV)", ""))
  else:
    rows.append(page.extract_text().replace("\n", " "))
```

```python
len("TABLE 2Nuclear Binding Energies A Z BE (MeV)  A Z BE (MeV)  A Z BE (MeV)  A Z BE (MeV)")
```

```python
cleaned_rows = []
for row in rows:
  row = row.replace("TABLE 2Nuclear Binding Energies. A Z BE (MeV)  A Z BE (MeV)  A Z BE (MeV)  A Z BE (MeV) ", "").replace("**", "").replace("*","").lstrip().rstrip().replace(" Y", "Y")
  cleaned_rows.append(row)
```

```python
# cleaned_rows
```

```python
elements = []
binding_energies = []

for row in cleaned_rows:
  lst = row.split(" ")
  cleaned_lst = []
  for element in lst:
    if element == '':
      continue
    cleaned_lst.append(element)
    
  el, be = cleaned_lst[::2], cleaned_lst[1::2]
  elements += el
  binding_energies += be

elements, binding_energies
```

```python
dataframe = pd.DataFrame(list(zip(elements, binding_energies)),columns=['element', 'binding_energy'])
```
